```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Jan22 = pd.read_excel('Jan22.xlsx')
Feb22 = pd.read_excel('Feb22.xlsx')
Mar22 = pd.read_excel('Mar22.xlsx')
April22 = pd.read_excel('April22.xlsx')
Mei22 = pd.read_excel('Mei22.xlsx')
Jun22 = pd.read_excel('Jun22.xlsx')
Jul22 = pd.read_excel('Jul22.xlsx')
Agu22 = pd.read_excel('Agu22.xlsx')
Sep22 = pd.read_excel('Sep22.xlsx')
Okt22 = pd.read_excel('Okt22.xlsx')
Nov22 = pd.read_excel('Nov22.xlsx')
```

```python
Data_covid_kecamatan = pd.concat([Jan22, Feb22, Mar22, April22, Mei22, Jun22, Jul22, Agu22, Sep22, Okt22, Nov22])
```

```python
Data_covid_kecamatan.groupby('nama_kecamatan')['POSITIF'].sum()
```

```python
Jumlah_positif_covid_kecamatan = Data_covid_kecamatan.groupby('nama_kecamatan')['POSITIF'].sum()
```

```python
fig = Data_covid_kecamatan.groupby('nama_kecamatan')['POSITIF'].sum().plot(kind='bar', figsize=(17,10), title='Cov-Positif Januari 2022 - November 2022 \n Prov. DKI Jakarta',xlabel='Kecamatan', ylabel='Jumlah')
#plt.savefig('Data_covid_kecamatan.jpg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
Data_covid_kecamatan.to_excel("Data_forecast_covid2022.xlsx", index=False)
```
