```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Jan21 = pd.read_excel('Jan21.xlsx')
Feb21 = pd.read_excel('Feb21.xlsx')
Mar21 = pd.read_excel('Mar21.xlsx')
Apr21 = pd.read_excel('Apr21.xlsx')
Mei21 = pd.read_excel('Mei21.xlsx')
Jun21 = pd.read_excel('Jun21.xlsx')
Jul21 = pd.read_excel('Jul21.xlsx')
Agu21 = pd.read_excel('Agu21.xlsx')
Sep21 = pd.read_excel('Sep21.xlsx')
Okt21 = pd.read_excel('Okt21.xlsx')
Nov21 = pd.read_excel('Nov21.xlsx')
Des21 = pd.read_excel('Des21.xlsx')
```

```python
Data_covid_kecamatan = pd.concat([Jan21, Feb21, Mar21, Apr21, Mei21, Jun21, Jul21, Agu21, Sep21, Okt21 ,Nov21, Des21])
```

```python
Data_covid_kecamatan.groupby('nama_kecamatan')['POSITIF'].sum()
```

```python
Jumlah_positif_covid_kecamatan = Data_covid_kecamatan.groupby('nama_kecamatan')['POSITIF'].sum()
```

```python
fig = Data_covid_kecamatan.groupby('nama_kecamatan')['POSITIF'].sum().plot(kind='bar', figsize=(17,10), title='Cov-Positif Januari 2021 - Desember 2021 \n Prov. DKI Jakarta',xlabel='Kecamatan', ylabel='Jumlah')
#plt.savefig('Data_covid_kecamatan.jpg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
Data_covid_kecamatan.to_excel("Data_forecast_covid2021.xlsx", index=False)
```
