```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Mar20 = pd.read_excel('Mar20.xlsx')
Apr20 = pd.read_excel('Apr20.xlsx')
Mei20 = pd.read_excel('Mei20.xlsx')
Jun20 = pd.read_excel('Jun20.xlsx')
Jul20 = pd.read_excel('Jul20.xlsx')
Agu20 = pd.read_excel('Agu20.xlsx')
Sep20 = pd.read_excel('Sep20.xlsx')
Okt20 = pd.read_excel('Okt20.xlsx')
Nov20 = pd.read_excel('Nov20.xlsx')
Des20 = pd.read_excel('Des20.xlsx')
```

```python
Data_covid_kecamatan = pd.concat([Mar20, Apr20, Mei20, Jun20, Jul20, Agu20, Sep20, Okt20, Nov20, Des20])
```

```python
Data_covid_kecamatan.groupby('nama_kecamatan')['POSITIF'].sum()
```

```python
Jumlah_positif_covid_kecamatan = Data_covid_kecamatan.groupby('nama_kecamatan')['POSITIF'].sum()
```

```python
fig = Data_covid_kecamatan.groupby('nama_kecamatan')['POSITIF'].sum().plot(kind='bar', figsize=(17,10), title='Cov-Positif Maret 2020 - Desember 2020 \n Prov. DKI Jakarta',xlabel='Kecamatan', ylabel='Jumlah')
#plt.savefig('Data_covid_kecamatan.jpg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
Data_covid_kecamatan.to_excel("Data_forecast_covid2020.xlsx", index=False)
```
